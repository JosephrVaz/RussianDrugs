```python
def translate_text(text):
  translation = translator.translate(text, src='ru', dest='en')
  return translation.text
```

```python
import spacy

# Load the Russian NLP model
nlp = spacy.load("ru_core_news_sm")

# Now you can use the 'nlp' object to process Russian text:
text = "Это пример русского текста." # This is an example of Russian text.
doc = nlp(text)

# Print tokens and their lemmas
for token in doc:
    print( token.lemma_ ,translate_text(token.lemma_))
```

```python
def preprocess(text):
  doc = nlp(text)
  tokens = [token.lemma_ for token in doc if
            not token.is_stop and not token.is_punct and not token.is_digit]
  return ' '.join(tokens)
```

```python
df['content_preprocess'] = df['content'].apply(preprocess)
```

```python
def tokenize_text(text):
  doc = nlp(text)
  tokens = [token.text for token in doc]
  return tokens
```

```python

pipeline = Pipeline([
    ('tfidf', TfidfVectorizer(tokenizer=tokenize_text)),
    ('classifier', RandomForestClassifier(n_estimators=100, max_depth=10))
])
```

```python
def hyperparamter_ml(X_train, y_train, X_test, y_test, n_estimators_list, max_depth_list):
  best_acc = 0
  best_params = {}
  for n_estimators in n_estimators_list:
    for max_depth in max_depth_list:
      pipeline = Pipeline([
          ('tfidf', TfidfVectorizer(tokenizer=tokenize_text)),
          ('classifier', RandomForestClassifier(n_estimators=n_estimators, max_depth=max_depth))
      ])
      pipeline.fit(X_train, y_train)
      preds = pipeline.predict(X_test)
      acc = accuracy_score(y_test, preds)
      if acc > best_acc:
        best_acc = acc
        best_params = {"n_estimators": n_estimators, "max_depth": max_depth, "accuracy": acc}
  return best_params
```

```python
import spacy
from spacy.pipeline.textcat import TextCategorizer
from spacy.training.example import Example
from spacy.util import minibatch
from sklearn.model_selection import train_test_split
from spacy.pipeline.textcat import Config, single_label_cnn_config

config = Config().from_str(single_label_cnn_config)
nlp2 = spacy.load("ru_core_news_md")
textcat = nlp2.add_pipe("textcat", last=True, config={
  'model': {'@architectures': 'spacy.TextCatReduce.v1',
  'exclusive_classes': True,
  'use_reduce_first': False,
  'use_reduce_last': False,
  'use_reduce_max': False,
  'use_reduce_mean': True,
  'tok2vec': {'@architectures': 'spacy.HashEmbedCNN.v2',
   'pretrained_vectors': None,
   'width': 96,
   'depth': 4,
   'embed_size': 4500,
   'window_size': 1,
   'maxout_pieces': 3,
   'subword_features': True}}

})
textcat.add_label("POSITIVE")
textcat.add_label("NEGATIVE")
```

```python
def df_to_spacy_format(df):
    new_data = []
    for index, row in df.iterrows():
        text = row['content_preprocess']
        label = row['sentiment']
        if label == 1:
            cats_dict = {"cats": {"POSITIVE": 1, "NEGATIVE": 0}}
        elif label == -1:
            cats_dict = {"cats": {"POSITIVE": 0, "NEGATIVE": 1}}


        new_data.append((text, cats_dict))
    return new_data
```

```python
df_spacy = df_to_spacy_format(df)
```

```python
df_spacy
```

```python
import random
random.shuffle(df_spacy)
shuffled_train = df_spacy[:int(len(df)*0.7)]
shuffled_test = df_spacy[int(len(df)*0.7):]
```

```python

def train_model(nlp, train_data, epochs=10, batch_size=32):
    """Trains the spaCy text classification model."""

    optimizer = nlp.initialize()
    losses = {}
    for epoch in range(epochs):
        random.shuffle(train_data)
        batches = minibatch(train_data, size=batch_size)
        for batch in batches:
            examples = [Example.from_dict(nlp.make_doc(text), annotations) for text, annotations in batch]
            nlp.update(examples, drop=None, losses=losses, sgd=optimizer)
        print(f"Epoch {epoch + 1}: Loss - {losses['textcat']}")

    return nlp  # Return the trained model
```

```python
nlp_trained = train_model(nlp2, shuffled_train)
```

```python
def find_marked_categories(cat_dict):
    marked_categories = [category for category, value in cat_dict['cats'].items() if value == 1]
    return marked_categories[0]

def predict_and_evaluate(model, test_data):
    correct_predictions = 0
    predictions = []

    for text, true_labels in test_data:
        doc = model(text)
        prediction = doc.cats

        highest_category = max(prediction, key=prediction.get)
        if find_marked_categories(true_labels) == highest_category:
            correct_predictions += 1

    return correct_predictions/len(test_data)
```

```python
train_acc = predict_and_evaluate(nlp_trained, shuffled_train)
test_acc = predict_and_evaluate(nlp_trained, shuffled_test)

print(f"Train accuracy: {round(train_acc*100)}")
print(f"Test accuracy: {round(test_acc*100)}")
```

This suggests model is overfitted

```python
def hyperparameters(nlp, train_data, epochs=10):
    batches =[4,8,16,32]
    dropouts = [0.2,0.3,0.4,0.5]
    best_acc = 0
    best_params = {}
    losses = {}
    for batch_size in batches: # This loop iterates through batch_size values
      for dropout in dropouts:
        # Changed 'batch' to 'train_data' to use the training data for initial examples
        examples = [Example.from_dict(nlp.make_doc(text), annotations) for text, annotations in train_data]
        optimizer = nlp.initialize(lambda: examples)
        for epoch in range(epochs):
          # The 'batch' variable is correctly assigned here in the inner loop
          batches_inner = minibatch(train_data, size=batch_size)
          for batch in batches_inner: # Iterating through mini-batches
            examples = [Example.from_dict(nlp.make_doc(text), annotations) for text, annotations in batch]
            nlp.update(examples, drop=dropout, losses=losses, sgd=optimizer)

        train_acc = predict_and_evaluate(nlp, shuffled_train)
        test_acc = predict_and_evaluate(nlp, shuffled_test)

        if test_acc > best_acc:
          best_acc = test_acc
          best_params = {"batch_size": batch_size, "dropout": dropout, "test_acc": test_acc, "train_acc": train_acc}

    return best_params
```

```python
print(hyperparameters(nlp2, shuffled_train))
```

Slighty improved with dropout
